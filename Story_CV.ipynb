# **Image Gen Code**

```python
from PIL import Image
import torch
from diffusers import DiffusionPipeline
import os

# Load the diffusion pipeline
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")

# Define a list of prompts for different frames of the video
prompts = [
    "Harry Potter playing football, about to kick the ball",
    "Harry Potter kicking the football",
    "Harry Potter just after kicking the football",
    "Harry Potter watching the football fly"
]

# Create directory to store the images
output_dir = "harry_potter_football_frames"
os.makedirs(output_dir, exist_ok=True)

# Generate images for each prompt
images = []
for i, prompt in enumerate(prompts):
    image = pipe(prompt=prompt).images[0]
    image.save(f"{output_dir}/frame_{i}.png")
    images.append(image)

# Display first image to check
images[0]
```

# **Video Gen Code**

```python
import cv2
import os

# Path where the generated frames are stored
image_folder = 'harry_potter_football_frames'
video_name = 'harry_potter_football.mp4'

# Fetch all images from the folder
images = [img for img in sorted(os.listdir(image_folder)) if img.endswith(".png")]

# Get the dimensions of the first image to set video size
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

# Define the video codec and initialize video writer
video = cv2.VideoWriter(video_name, cv2.VideoWriter_fourcc(*'mp4v'), 1, (width, height))

# Write each image to the video
for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

# Release the video writer
cv2.destroyAllWindows()
video.release()

print("Video created successfully!")
```

# **Cartoon Video**

```python
from PIL import Image
import torch
from diffusers import DiffusionPipeline
import os
import cv2

# Load the diffusion pipeline for cartoon-style generation
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")  # Switch to "cpu" if GPU is unavailable

# Define a list of prompts for different frames of the video
prompts = [
    "A cartoon boy running through a magical forest, digital art style",
    "A castle in the background with glowing stars in the sky",
    "A talking tree with a friendly face in a magical forest, cartoon style",
    "The boy meeting a fairy near the castle, colorful cartoon",
    "Flying dragons over a cartoon village, sunset scene",
    "The boy and the fairy flying over the mountains, cartoon animation style"
]

# Create directory to store the images
output_dir = "cartoon_story_frames"
os.makedirs(output_dir, exist_ok=True)

# Generate images for each prompt
images = []
for i, prompt in enumerate(prompts):
    image = pipe(prompt=prompt).images[0]
    image.save(f"{output_dir}/frame_{i}.png")
    images.append(image)

# Display first image to check
images[0].show()

# Convert generated images to video
image_folder = 'cartoon_story_frames'
video_name = 'cartoon_story_video.mp4'

# Fetch all images from the folder
images = [img for img in sorted(os.listdir(image_folder)) if img.endswith(".png")]

# Get the dimensions of the first image to set video size
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

# Define the video codec and initialize video writer
fps = 1  # Adjust fps for slower display, set it to higher for actual 24fps video creation
video = cv2.VideoWriter(video_name, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Write each image to the video
for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

# Release the video writer
cv2.destroyAllWindows()
video.release()

print("Video created successfully!")
```
